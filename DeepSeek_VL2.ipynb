```python
import gc
import traceback
import re
import os
import time
from io import BytesIO
from PIL import Image
import requests
import torch

from deepseek_vl2.models import DeepseekVLV2Processor, DeepseekVLV2ForCausalLM

MODEL_PATH = "deepseek-ai/deepseek-vl2-tiny"
IMAGE_URL = "https://hips.hearstapps.com/hmg-prod/images/most-popular-dog-breeds-bracco-italiano-68096ed33c726.jpg"
MAX_NEW_TOKENS = 256

# Set environment variables for better performance
os.environ["TOKENIZERS_PARALLELISM"] = "false"
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"

def print_device_info():
    print(f"PyTorch: {torch.__version__}")
    print(f"CUDA available: {torch.cuda.is_available()}")
    if torch.cuda.is_available():
        prop = torch.cuda.get_device_properties(0)
        print(f"CUDA device: {prop.name}")
        print(f"CUDA capability: {prop.major}.{prop.minor}")
        print(f"Total CUDA memory (GB): {prop.total_memory / (1024**3):.2f}")

def load_image(url):
    resp = requests.get(url, timeout=20)
    resp.raise_for_status()
    img = Image.open(BytesIO(resp.content)).convert("RGB")
    return img

def build_conversation():
    return [
        {"role": "user", "content": "What do you see in this image? <image>"},
        {"role": "assistant", "content": ""}
    ]

def prepare_processor_inputs(processor, image):
    conv = build_conversation()
    batch = processor(
        conversations=conv,
        images=[image],
        sft_format="deepseek",
        return_tensors="pt"
    )
    return batch

def collator_to_valid_inputs(batch):
    return {
        "input_ids": getattr(batch, "input_ids", None),
        "attention_mask": getattr(batch, "attention_mask", None),
        "images": getattr(batch, "images", None),
        "images_seq_mask": getattr(batch, "images_seq_mask", None),
        "images_spatial_crop": getattr(batch, "images_spatial_crop", None),
    }

def cast_and_move_inputs(valid_inputs, model):
    device = next(model.parameters()).device
    model_dtype = model.dtype
    for k, v in list(valid_inputs.items()):
        if v is None:
            continue
        if not isinstance(v, torch.Tensor):
            continue
        if v.dtype in (torch.int64, torch.int32, torch.bool):
            valid_inputs[k] = v.to(device=device)
        else:
            try:
                valid_inputs[k] = v.to(device=device, dtype=model_dtype)
            except Exception as e:
                print(f"Warning: couldn't cast '{k}' to {model_dtype}: {e}. Moving without casting.")
                valid_inputs[k] = v.to(device=device)
    return valid_inputs

def clean_generated_text(generated: str, prompt_text: str) -> str:
    """
    Clean model output by:
    - removing echoed prompt if present
    - removing bounding box tokens like [[...]]
    - removing stray glyphs and repeated noise
    - removing repetitive sentences or phrases with normalization
    - handling DeepSeek-specific repetition patterns
    """
    if not generated:
        return ""

    text = generated

    # Remove echoed prompt
    if prompt_text and prompt_text in text:
        text = text.split(prompt_text)[-1]

    # Remove bounding boxes entirely
    text = re.sub(r'\[\[[^\]]+\]\]', '', text)

    # Remove leading role separators
    text = re.sub(r'^\s*[:\-–—\s]+', '', text)

    # Remove repeated stray glyphs
    text = re.sub(r'(?:▽\s*){2,}', ' ', text)
    text = re.sub(r'(?:\d+:\s*▽\s*){2,}', '', text)
    text = re.sub(r'([^\w\s\[\]\.,\-])\1{2,}', r'\1', text)

    # Handle DeepSeek-specific repetition patterns
    # Remove repeated sentences with slight variations
    sentences = re.split(r'(?<=[.!?])\s+', text.strip())

    # More sophisticated deduplication that handles similar sentences
    seen_sentences = set()
    unique_sentences = []

    for sentence in sentences:
        if not sentence.strip():
            continue

        # Normalize sentence for comparison
        normalized = sentence.strip().lower()
        normalized = re.sub(r'[^\w\s]', '', normalized)  # Remove punctuation
        normalized = re.sub(r'\s+', ' ', normalized).strip()

        # Check if this is a substantial duplicate
        is_duplicate = False
        for seen in seen_sentences:
            if normalized in seen or seen in normalized or \
               (len(normalized) > 10 and len(seen) > 10 and
                (normalized.startswith(seen[:10]) or seen.startswith(normalized[:10]))):
                is_duplicate = True
                break

        if not is_duplicate:
            seen_sentences.add(normalized)
            unique_sentences.append(sentence.strip())

    # Rejoin sentences
    text = ' '.join(unique_sentences)

    # Remove any remaining repetitive phrases
    text = re.sub(r'(\b\w+\b)(?:\s+\1\b)+', r'\1', text)

    # Collapse spaces/newlines
    text = re.sub(r'\n{2,}', '\n', text)
    text = re.sub(r'[ \t]{2,}', ' ', text)

    # Clean edges
    text = text.strip()
    text = re.sub(r'^[\s\:\-–—\|]+', '', text)
    text = re.sub(r'[\s\:\-–—\|]+$', '', text)

    # Remove empty or non-text content
    if re.fullmatch(r'[\W_]+', text):
        return ""

    return text.strip()

class EfficientVL2Processor:
    """Efficient processor with memory optimization and caching"""
    def __init__(self):
        self.processor = None
        self.model = None

    def initialize(self, dtype=torch.float16):
        """Initialize with memory-efficient settings"""
        if self.processor is None:
            self.processor = DeepseekVLV2Processor.from_pretrained(MODEL_PATH)

        if self.model is None:
            self.model = DeepseekVLV2ForCausalLM.from_pretrained(
                MODEL_PATH,
                torch_dtype=dtype,
                device_map="auto",
                low_cpu_mem_usage=True,
            )
            self.model.eval()

        return self.model

def try_generate_with_dtype(dtype):
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        torch.cuda.reset_peak_memory_stats()

    print(f"\n== Loading model with dtype={dtype} ==")

    # Use efficient processor with memory optimization
    efficient_processor = EfficientVL2Processor()
    model = efficient_processor.initialize(dtype)
    processor = efficient_processor.processor

    img = load_image(IMAGE_URL)

    # Process inputs with minimal memory footprint using inference_mode
    with torch.inference_mode():
        batch = prepare_processor_inputs(processor, img)
        valid_inputs = collator_to_valid_inputs(batch)
        valid_inputs = cast_and_move_inputs(valid_inputs, model)

    # Memory monitoring
    if torch.cuda.is_available():
        print(f"Memory allocated: {torch.cuda.memory_allocated() / 1024**2:.2f} MB")
        print(f"Memory reserved: {torch.cuda.memory_reserved() / 1024**2:.2f} MB")

    if valid_inputs.get("images", None) is None:
        raise RuntimeError("Expected 'images' tensor but got None from the processor.")
    if valid_inputs.get("images_spatial_crop", None) is None:
        raise RuntimeError("Expected 'images_spatial_crop' tensor but got None from the processor.")

    print("All inputs moved to device and cast to model dtype:", model.dtype)
    print("input_ids device/dtype:", valid_inputs["input_ids"].device, valid_inputs["input_ids"].dtype)
    print("images device/dtype:", valid_inputs["images"].device, valid_inputs["images"].dtype)
    print("images_spatial_crop device/dtype:", valid_inputs["images_spatial_crop"].device, valid_inputs["images_spatial_crop"].dtype)

    # Optimized generation with better parameters
    with torch.no_grad():
        output = model.generate(
            **valid_inputs,
            max_new_tokens=MAX_NEW_TOKENS,
            do_sample=False,
            repetition_penalty=1.1,  # Slightly lower penalty for better quality
            no_repeat_ngram_size=2,  # Prevent repeating 2-grams
            pad_token_id=processor.tokenizer.eos_token_id,
            use_cache=True,  # Enable KV caching for speed
        )

    generated = processor.tokenizer.batch_decode(output, skip_special_tokens=True)[0]
    prompt_text = "What do you see in this image? <image>"
    cleaned = clean_generated_text(generated, prompt_text)

    # Cleanup but keep model cached for potential reuse
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    return cleaned

def load_quantized_model():
    """Load model with 4-bit quantization for maximum efficiency"""
    try:
        from transformers import BitsAndBytesConfig

        quantization_config = BitsAndBytesConfig(
            load_in_4bit=True,
            bnb_4bit_compute_dtype=torch.float16,
            bnb_4bit_use_double_quant=True,
            bnb_4bit_quant_type="nf4",
        )

        model = DeepseekVLV2ForCausalLM.from_pretrained(
            MODEL_PATH,
            quantization_config=quantization_config,
            device_map="auto",
            low_cpu_mem_usage=True,
        )
        return model
    except ImportError:
        print("BitsAndBytes not available, falling back to regular loading")
        return None

def main():
    print_device_info()

    # Set PyTorch optimizations
    torch.set_float32_matmul_precision('high')
    if hasattr(torch.backends, 'cudnn'):
        torch.backends.cudnn.benchmark = True

    start_time = time.time()
    dtype_candidates = [torch.float16, torch.float32]
    tried = []
    last_exc = None

    for dtype in dtype_candidates:
        tried.append(dtype)
        try:
            result = try_generate_with_dtype(dtype)
            end_time = time.time()

            print(f"\n=== SUCCESS (dtype={dtype}) ===")
            print(f"Processing time: {end_time - start_time:.2f} seconds")

            # Memory stats
            if torch.cuda.is_available():
                peak_memory = torch.cuda.max_memory_allocated() / 1024**2
                print(f"Peak memory usage: {peak_memory:.2f} MB")
                torch.cuda.reset_peak_memory_stats()

            print("\nGenerated output:\n", result)
            return

        except Exception as e:
            last_exc = e
            print(f"\nAttempt with dtype={dtype} failed:")
            traceback.print_exc()
            print(f"Falling back. Tried so far: {tried}\n")

    # Try quantization as last resort
    print("Trying quantized model as last resort...")
    try:
        quantized_model = load_quantized_model()
        if quantized_model:
            # Reuse the existing processor
            processor = DeepseekVLV2Processor.from_pretrained(MODEL_PATH)
            quantized_model.eval()

            img = load_image(IMAGE_URL)
            batch = prepare_processor_inputs(processor, img)
            valid_inputs = collator_to_valid_inputs(batch)
            valid_inputs = cast_and_move_inputs(valid_inputs, quantized_model)

            with torch.no_grad():
                output = quantized_model.generate(
                    **valid_inputs,
                    max_new_tokens=MAX_NEW_TOKENS,
                    do_sample=False,
                )

            generated = processor.tokenizer.batch_decode(output, skip_special_tokens=True)[0]
            cleaned = clean_generated_text(generated, "What do you see in this image? <image>")

            print(f"\n=== SUCCESS (Quantized) ===")
            print("Generated output:\n", cleaned)
            return
    except Exception as e:
        print("Quantized model also failed:", e)

    raise RuntimeError(f"All dtype attempts failed ({tried}). Last error: {last_exc}")

if __name__ == "__main__":
    main()
```

```python
# Add these imports at the top of your Colab notebook
from flask import Flask, request, jsonify
from flask_ngrok import run_with_ngrok
import threading
import requests
from PIL import Image
import io
import base64

# Add this after your existing imports and functions
app = Flask(__name__)

# Initialize the model and processor (reuse your existing code)
def initialize_model():
    # Your existing model initialization code here
    processor = DeepseekVLV2Processor.from_pretrained(MODEL_PATH)
    model = DeepseekVLV2ForCausalLM.from_pretrained(
        MODEL_PATH,
        torch_dtype=torch.float16,
        device_map="auto",
        low_cpu_mem_usage=True,
    )
    model.eval()
    return processor, model

# Global variables for model and processor
processor, model = initialize_model()

@app.route('/process', methods=['POST'])
def process_file():
    try:
        if 'file' not in request.files:
            return jsonify({'error': 'No file provided'})

        file = request.files['file']
        filename = file.filename.lower()

        if filename.endswith(('.png', '.jpg', '.jpeg', '.gif')):
            # Process image
            image = Image.open(file.stream).convert('RGB')
            return process_image(image)

        elif filename.endswith('.pdf'):
            # Extract text from PDF and process as text
            import PyPDF2
            pdf_reader = PyPDF2.PdfReader(file.stream)
            text = ""
            for page in pdf_reader.pages:
                text += page.extract_text() + "\n"
            return process_text(text)

        else:
            return jsonify({'error': 'Unsupported file type'})

    except Exception as e:
        return jsonify({'error': f'Processing error: {str(e)}'})

@app.route('/process_text', methods=['POST'])
def process_text_endpoint():
    try:
        text = request.form.get('text', '')
        if not text:
            return jsonify({'error': 'No text provided'})

        return process_text(text)
    except Exception as e:
        return jsonify({'error': f'Text processing error: {str(e)}'})

def process_image(image):
    try:
        # Your existing image processing code here
        conv = [{"role": "user", "content": "What do you see in this image? <image>"},
               {"role": "assistant", "content": ""}]

        batch = processor(
            conversations=conv,
            images=[image],
            sft_format="deepseek",
            return_tensors="pt"
        )

        valid_inputs = collator_to_valid_inputs(batch)
        valid_inputs = cast_and_move_inputs(valid_inputs, model)

        with torch.no_grad():
            output = model.generate(
                **valid_inputs,
                max_new_tokens=MAX_NEW_TOKENS,
                do_sample=False,
                repetition_penalty=1.1,
                no_repeat_ngram_size=2,
                pad_token_id=processor.tokenizer.eos_token_id,
                use_cache=True,
            )

        generated = processor.tokenizer.batch_decode(output, skip_special_tokens=True)[0]
        cleaned = clean_generated_text(generated, "What do you see in this image? <image>")

        return jsonify({'result': cleaned})

    except Exception as e:
        return jsonify({'error': f'Image processing error: {str(e)}'})

def process_text(text):
    try:
        # Process text input
        conv = [{"role": "user", "content": text},
               {"role": "assistant", "content": ""}]

        batch = processor(
            conversations=conv,
            images=None,  # No image for text processing
            sft_format="deepseek",
            return_tensors="pt"
        )

        valid_inputs = collator_to_valid_inputs(batch)
        valid_inputs = cast_and_move_inputs(valid_inputs, model)

        with torch.no_grad():
            output = model.generate(
                **valid_inputs,
                max_new_tokens=MAX_NEW_TOKENS,
                do_sample=False,
                repetition_penalty=1.1,
                no_repeat_ngram_size=2,
                pad_token_id=processor.tokenizer.eos_token_id,
                use_cache=True,
            )

        generated = processor.tokenizer.batch_decode(output, skip_special_tokens=True)[0]
        cleaned = clean_generated_text(generated, text)

        return jsonify({'result': cleaned})

    except Exception as e:
        return jsonify({'error': f'Text processing error: {str(e)}'})

# Run the Flask server in Colab
def run_flask():
    run_with_ngrok(app)  # Start ngrok when app is run
    app.run()

# Start the Flask server in a separate thread
flask_thread = threading.Thread(target=run_flask)
flask_thread.daemon = True
flask_thread.start()

print("Server is running! Check the output for the ngrok URL.")
```
